```python
from qiskit import transpile,QuantumRegister,ClassicalRegister,QuantumCircuit

# 创建包含1个量子比特的量子寄存器
qr=QuantumRegister(1)
# 创建包含1个经典比特的经典寄存器
cr=ClassicalRegister(1)
# 创建量子电路qc，包含量子寄存器qr和经典寄存器cr
qc=QuantumCircuit(qr,cr)
# 添加Hadamard门到量子电路qc的第0个量子比特
qc.h(0)
# 测量量子电路qc的第0个量子比特，并将结果存储到第0个经典比特
qc.measure(0,0)
# 如果第0个经典比特的值为0，则添加X门到量子电路qc的第0个量子比特
qc.x(0).c_if(cr,0)
# 再次测量量子电路qc的第0个量子比特，并将结果存储到第0个经典比特
qc.measure(0,0)
# 绘制量子电路qc
qc.draw('mpl')
```

```python
from qiskit.providers.basic_provider import BasicSimulator
backend=BasicSimulator()
tqc=transpile(qc,backend)
counts=backend.run(tqc).result().get_counts()

print(counts)
```

```python
from qiskit import QuantumCircuit
# 创建一个有12个量子比特的量子电路
qc=QuantumCircuit(12)
# 建立量子纠缠
for idx in range(5):
    qc.h(idx)
    # 如果控制量子比特处于∣0⟩态，则目标量子比特的状态不变
    # 如果控制量子比特处于∣1⟩态，则目标量子比特的状态翻转
    qc.cx(idx,idx+5)
qc.cx(1,7)
qc.x(8)
qc.cx(1,9)
qc.x(7)
qc.cx(1,11)
qc.swap(6,11)
qc.swap(6,9)
qc.swap(6,10)
qc.x(6)
qc.draw('mpl')
```
